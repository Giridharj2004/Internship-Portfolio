```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\data analytics task 1\dirty_cafe_sales.csv")
df.isnull().sum()
df.head()
```

```python
df.replace(['ERROR','UNKNOWN'],np.nan,inplace=True)
df.isnull().sum()
```

```python
df['Quantity']=pd.to_numeric(df['Quantity'],errors='coerce')
df['Price Per Unit']=pd.to_numeric(df['Price Per Unit'],errors='coerce')
df['Total Spent']=pd.to_numeric(df['Total Spent'],errors='coerce')
df['Transaction Date']=pd.to_datetime(df['Transaction Date'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Item'].fillna(df['Item'].mode()[0], inplace=True)
```

```python
df['Quantity'].fillna(df['Quantity'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Price Per Unit'].fillna(df['Price Per Unit'].mean(),inplace=True)
```

```python
df['Total Spent'].fillna(df['Total Spent'].mean(),inplace=True)
```

```python
df['Payment Method'].fillna('Unknown',inplace=True)
df['Location'].fillna('Unknown',inplace=True)
df['Transaction Date'].fillna(df['Transaction Date'].mode()[0],inplace=True)

```

```python
df['Total Spent']=df['Quantity']*df['Price Per Unit']
```

```python
df['Calculated Total']=df['Quantity']*df['Price Per Unit']
```

```python
df['Day'] = df['Transaction Date'].dt.day_name()
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
